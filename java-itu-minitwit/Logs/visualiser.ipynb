```python
from matplotlib import pyplot as plt
```

```python
def get_lines(path):
    return open(path+".txt", 'r').readlines()

def parse_start(lines):
    values = []
    dates = []
    for (line_number, line) in enumerate(lines):
        split = line.replace(" \n", "").split(" - ")
        values.append(float(split[0]))
        dates.append(split[1])
    return (values, dates)

def printFiles(filetype, dates):
    lines = []
    ammounts = []
    for date in dates:
        file = get_lines(filetype+date)
        lines = lines + file
        ammounts.append(len(lines))
    return (lines, ammounts)

def plotdata(lines, ammounts, xaxisName, yaxisName):
    values, dates = parse_start(lines)
    maxValue = max(values)
    for a in ammounts:
        plt.vlines(a, 0, maxValue, linestyles="-", colors="black")
    plt.xlabel(xaxisName)
    plt.ylabel(yaxisName)
    plt.plot(values)

def print_cpu(dates):
    lines, ammounts = printFiles('CPULoadEachMinute-',dates)
    plotdata(lines, ammounts, "time", "cpu")

def print_users(dates):
    lines, ammounts = printFiles('numberOfUsers-',dates)
    plotdata(lines, ammounts, "time", "users")

def print_followers(dates):
    lines, ammounts = printFiles('numberOfFollowers-',dates)
    plotdata(lines, ammounts, "time", "average followers")

def print_response_time(dates):
    lines, ammounts = printFiles('responseTimeFrontPage-',dates)
    plotdata(lines, ammounts, "time", "response time[ns]")
```

```python
print_cpu(['21-2-5' ])
```

```python
print_users(['21-2-5'])
```

```python
print_followers(['21-2-5'])
```

```python
print_response_time(['21-2-5'])
```
