```python
from matplotlib import pyplot as plt
```

```python
def get_lines(path):
    return open(path, 'r').readlines()

def parse_start(lines):
    values = []
    dates = []
    for (line_number, line) in enumerate(lines):
        split = line.replace(" \n", "").split(" - ")
        values.append(float(split[0]))
        dates.append(split[1])
    return (values, dates)
```

```python
values, dates = parse_start(get_lines('CPULoadEachMinute-21-2-5.txt'))
plt.plot(values)
```

```python
values, dates = parse_start(get_lines('numberOfFollowers-21-2-5.txt'))
plt.plot(values)
```
